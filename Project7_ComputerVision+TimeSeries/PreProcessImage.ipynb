```python
import re
```

```python
def getDate(ls):
  date=re.search(r'(\d{1}|{2}[.|,]\w+(.|,){1}\d{4})',' '.join(ls))
  time=re.search(r'(\d+[.|,|:]\d+[.|,|:]\d{2})',' '.join(ls))
  if date and time:
    return date.group(1)+' '+time.group(1).replace(',','.').replace('.',':')
```
