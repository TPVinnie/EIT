```python
import os
import easyocr
import pandas as pd

img_path = '/content/images/'  # Path to image folder
files = os.listdir(img_path)

reader = easyocr.Reader(['en'])  # Language(s) for text recognition

data = []  # List to store image IDs and corresponding datetime

for file in files:
    # Read text from the image
    res = reader.readtext(img_path + file, detail=0)
    print(res)

    # Extract the image ID from the filename (without the extension)
    id = os.path.splitext(file)[0]

    # Convert the extracted text to datetime (assuming the datetime is in the last two elements of res list)
    if len(res) > 0:
        datetime_str = res[-2:]  # Assuming the datetime is in the last two elements of the res list
        data.append({'filename': id, 'datetime': datetime_str})

# Create a DataFrame from the collected data
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```
