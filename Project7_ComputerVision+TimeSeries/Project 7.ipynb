```python
from datetime import datetime
```

```python
def modif(dtime):
  dat=(eval(dtime)[0][:2]+'.'+eval(dtime)[0][-8:-5]+' '+eval(dtime)[0][-4:]).replace('z','2').replace('q','g').replace('t','g')
  tim=eval(dtime)[1].replace('.',':')
  try:
    dtime=dat+' '+tim
    #print(dtime)
    return datetime.strptime(dtime, '%d.%b %Y %H:%M:%S')
  except:
    pass
```
