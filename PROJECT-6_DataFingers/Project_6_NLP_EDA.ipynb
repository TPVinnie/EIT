```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import nltk
nltk.download('punkt')
import seaborn as sns
```

```python
from wordcloud import WordCloud
```

# Serious_Fun EDA

```python
df = pd.read_csv("/content/Serious_Fun_DATASET.csv")
```

```python
df.columns
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df['Serious(-1)_Fun(1)'] = df['Serious(-1)_Fun(1)'].astype(int)
```

```python
sentiment_count = df["Serious(-1)_Fun(1)"].value_counts()
```

```python
plt.pie(sentiment_count, labels = sentiment_count.index, autopct='%1.1f%%',shadow=True, startangle = 140)
plt.show()
```

```python
duplicate = df.duplicated().sum()
```

```python
duplicate
```

```python
# Calculate the length of each text entry (number of characters including spaces)
df['Text_Length'] = df['Text'].apply(lambda x: len(str(x)))

# Descriptive statistics of text length
text_length_stats = df['Text_Length'].describe()
print("Text Length Statistics:")
print(text_length_stats)

# Plot a histogram of text lengths with colored bins
num_bins = 100
baby_sky_color = '#89CFF0'  # Define the baby sky color
plt.figure(figsize=(10, 6))

n, bins, patches = plt.hist(df['Text_Length'], bins=num_bins)
bin_centers = 0.5 * (bins[:-1] + bins[1:])

# Assign the baby sky color to the patches
for value, patch in zip(bin_centers, patches):
    patch.set_facecolor(baby_sky_color)

plt.title('Text Length Distribution')
plt.xlabel('Text Length')
plt.ylabel('Frequency')

plt.show()
```

```python
# Tokenize the text into sentences
sentences = df['Text'].apply(nltk.sent_tokenize)

# Count the number of words in each sentence
word_counts = sentences.apply(lambda x: [len(nltk.word_tokenize(sentence)) for sentence in x])

# Calculate the average number of words per sentence
average_words_per_sentence = word_counts.apply(lambda x: sum(x) / len(x))

# Descriptive statistics of average words per sentence
average_words_stats = average_words_per_sentence.describe()
print("Average Words per Sentence Statistics:")
print(average_words_stats)

# Plot a histogram of average words per sentence
plt.figure(figsize=(10, 6))
plt.hist(average_words_per_sentence, bins=50)
plt.title('Average Words per Sentence Distribution')
plt.xlabel('Average Words per Sentence')
plt.ylabel('Frequency')
plt.show()
```

```python

df['word_count'] = df['Text'].apply(lambda x: len(str(x).split()))
plt.figure(figsize=(10,8))
sns.histplot(data=df, x='word_count')
plt.title('Word Count of sentence in Serious-Fun Data')
plt.xlabel('Word Count')
plt.ylabel('Sentence Count')
plt.show()
```

```python
# Filter the DataFrame for sentiment=1
df_sentiment1 = df[df['Serious(-1)_Fun(1)'] == -1]

# Combine all the text entries with sentiment=1 into a single string
all_text_sentiment1 = ' '.join(df_sentiment1['Text'])

# Generate word cloud based on word frequencies for sentiment=1
wordcloud_sentiment1 = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(all_text_sentiment1)

# Display the word cloud for sentiment=1 using matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_sentiment1, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud - Top 100 Words for Sentiment=1')
plt.show()
```

```python

# Calculate the length of each text entry (number of characters including spaces)
df2['Text_Length'] = df2['Text'].apply(lambda x: len(str(x)))

# Descriptive statistics of text length
text_length_stats = df2['Text_Length'].describe()
print("Text Length Statistics:")
print(text_length_stats)

# Plot a histogram of text lengths with colored bins
num_bins = 100
cmap = plt.cm.get_cmap('coolwarm')
plt.figure(figsize=(10, 6))

n, bins, patches = plt.hist(df2['Text_Length'], bins=num_bins)
bin_centers = 0.5 * (bins[:-1] + bins[1:])

# Assign colors to the patches
for value, patch in zip(bin_centers, patches):
    color = cmap((value - min(bin_centers)) / (max(bin_centers) - min(bin_centers)))
    patch.set_facecolor(color)

plt.title('Text Length Distribution')
plt.xlabel('Text Length')
plt.ylabel('Frequency')
sm = plt.cm.ScalarMappable(cmap=cmap)
sm.set_array([])
plt.colorbar(sm)

plt.show()

```
