```python
// static/script.js
function appendMessage(msg, sender) {
    var chatBox = document.getElementById("chat-box");
    var div = document.createElement("div");
    div.className = "chat-message " + sender;
    div.innerText = msg;
    chatBox.appendChild(div);
    chatBox.scrollTop = chatBox.scrollHeight;
}

function sendMessage() {
    var inputBox = document.getElementById("user-input-box");
    var userMessage = inputBox.value;
    appendMessage("You: " + userMessage, "user");
    inputBox.value = "";

    // Call the Flask backend to get the gift recommendation
    fetch('/recommend', {
        method: 'POST',
        headers: {
            'Content-Type': 'application/json'
        },
        body: JSON.stringify({ 'question': userMessage })
    })
    .then(response => response.json())
    .then(data => {
        var botResponse = data.recommendation;
        appendMessage("Bot: " + botResponse, "bot");
    })
    .catch(error => {
        console.error('Error:', error);
    });
}
```
