# Use the `IPython.display` module to render HTML and JavaScript directly in this notebook

```python
from IPython.display import display, HTML
import json

def chatbot_interface():
    html_code = '''
    <div class="chat-container">
        <div class="chat-header">
            <h1>Chatbot Interface</h1>
        </div>
        <div class="chat-box" id="chat-box">
            <div class="chat-message bot">Hello! How can I help you?</div>
        </div>
        <div class="user-input">
            <input type="text" id="user-input-box" placeholder="Type your message...">
            <button onclick="sendMessage()">Send</button>
        </div>
    </div>

    <script>
        function appendMessage(msg, sender) {
            var chatBox = document.getElementById("chat-box");
            var div = document.createElement("div");
            div.className = "chat-message " + sender;
            div.innerText = msg;
            chatBox.appendChild(div);
            chatBox.scrollTop = chatBox.scrollHeight;
        }

        function sendMessage() {
            var inputBox = document.getElementById("user-input-box");
            var userMessage = inputBox.value;
            appendMessage("You: " + userMessage, "user");
            inputBox.value = "";

            // Call Python function to get chatbot response
            var kernel = IPython.notebook.kernel;
            var command = 'get_chatbot_response("' + userMessage + '")';
            kernel.execute(command, {
                iopub: {
                    output: function(response) {
                        var botResponse = response.content.text.trim();
                        appendMessage("Bot: " + botResponse, "bot");
                    }
                }
            });
        }
    </script>
    '''

    # Display the chatbot interface
    display(HTML(html_code))

chatbot_interface()
```
